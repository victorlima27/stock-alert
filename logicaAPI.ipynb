````python
import requests, json
from datetime import datetime, timedelta

format_data = '%Y-%m-%dT%H:%M:%S'
````

````python
def busca_produtos(get):
    url_API = f'http://localhost:8080/api/{get}'
    
    response = requests.request("GET",url_API)
    if response.status_code == 200:
        resp = json.loads(response.text)
        return resp
````

````python
def dias_para_vencimento(entry_at, validade):
    data_entrada = datetime.strptime(entry_at, format_data)
    data_vencimento = data_entrada + validade
    agora = datetime.now()
    diferenca = data_vencimento - agora
    dias_restantes = diferenca.days
    return dias_restantes
````

````python
### Definição das validades de produtos ###
validade_legumes_verduras_crus = timedelta(days=5)
````

````python
# Get irá informar o tipo de endpoint utilizado
get = 'product-entries'

# Cria um objeto do response da API
lista_produtos = busca_produtos(get)
````

````python
produtos_vencidos = []
produtos_prox_vencimento = []
produtos_em_datas = []

for produto in lista_produtos:
    data_entrada = datetime.strptime(produto['entryAt'],format_data)
    dias_restantes = dias_para_vencimento(produto['entryAt'], validade_legumes_verduras_crus)
    print(produto['productName'])
    print(produto['quantity'])
    print(data_entrada)
    print(dias_restantes)

    produto_info = {
        'productName': produto['productName'],
        'quantity': produto['quantity'],
        'entryAt': data_entrada,
        'diasParaVencimento': dias_restantes
    }

    if dias_restantes <= 0:
        produtos_vencidos.append(produto_info)
    elif 1 <= dias_restantes <= validade_legumes_verduras_crus.days:
        produtos_prox_vencimento.append(produto_info)
        

    
````
